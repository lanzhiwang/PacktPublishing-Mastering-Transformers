```python
from datasets import load_dataset

imdb = load_dataset("imdb")
```

```python
imdb = imdb.shuffle()
```

```python
imdb_x = [x for x in imdb["train"][:1000]["text"]]
labels = [x for x in imdb["train"][:1000]["label"]]
import pandas as pd

pd.DataFrame(imdb_x, columns=["text"]).to_excel("imdb.xlsx", index=None)
```

```python
pd.read_excel("KHMER.xlsx")
```

```python
imdb_khmer = list(pd.read_excel("KHMER.xlsx").text)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y, khmer_train, khmer_test = train_test_split(
    imdb_x, labels, imdb_khmer, test_size=0.2, random_state=1
)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("stsb-xlm-r-multilingual")
```

```python
encoded_train = model.encode(train_x)
encoded_test = model.encode(test_x)
encoded_khmer_test = model.encode(khmer_test)
```

```python
import numpy as np

train_y = np.array(train_y)
test_y = np.array(test_y)
```

```python
import tensorflow as tf

input_ = tf.keras.layers.Input((768,))
classification = tf.keras.layers.Dense(1, activation="sigmoid")(input_)
classification_model = tf.keras.Model(input_, classification)
classification_model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer="Adam",
    metrics=["accuracy", "Precision", "Recall"],
)
```

```python
classification_model.fit(
    x=encoded_train, y=train_y, validation_data=(encoded_test, test_y), epochs=20
)
```

```python
classification_model.evaluate(x=encoded_khmer_test, y=test_y)
```
